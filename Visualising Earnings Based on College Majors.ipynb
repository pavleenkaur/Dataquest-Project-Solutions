# Comparing number of Men and Women in each Major

```python
#Evaluation of first 40 entries
fig, ax = plt.subplots(figsize=(15,8))
from numpy import arange
n = 40
bar_positions = arange(n) + 0.75         
bar_h1 = recent_grads["Men"].values.tolist()
bar_h1 = map(int,bar_h1[:n])
bar_h2 = recent_grads["Women"].values.tolist()
bar_h2 = map(int,bar_h2[:n])

tick_positions=range(1,n+1)
ax.set_xticks(tick_positions)
ax.bar(bar_positions,bar_h1,color="g",width=0.2,label="Men")
ax.bar(bar_positions+0.2,bar_h2,color="b",width=0.2,label="Women")

ax.legend(loc="upper right")
majors = recent_grads["Major"].values.tolist()
majors = majors[:n]
ax.set_xticklabels(majors, rotation=90)
plt.show()
```

## Analysis
### A surge of women is clearly visible in Nursing, and is minutely present in Computer Science and Environmental Engineering

# Comparing Median Salary for all Majors

```python
#Evaluation of first 50 entries
fig, ax = plt.subplots(figsize=(15,8))
n = 50
bar_positions = arange(n) + 0.75  

bar_h1 = recent_grads["Median"].values.tolist()
bar_height = map(int,bar_h1[:n])

tick_positions=range(1,n+1)
ax.set_xticks(tick_positions)
ax.bar(bar_positions,bar_height,color="r",width=0.2)

ax.legend(loc="upper right")
majors = recent_grads["Major"].values.tolist()
majors = majors[:n]
ax.set_xticklabels(majors, rotation=90)
plt.show()
```

## Analysis
### As can be observed, the top three salaried majors are Petroleum, Mining and Mineral followed by Metallurgical.
